## Data Collection
The crime dataset was collected from the open data portal of the "Organismo de Investigación Judicial" (OIJ) of Costa Rica [[1]](https://sitiooij.poder-judicial.go.cr/index.php/ayuda/servicios-policiales/servicios-a-organizaciones/indice-de-transparencia-del-sector-publico-costarricense/datos-abiertos)[[2]](https://sitiooij.poder-judicial.go.cr/index.php/ayuda/servicios-policiales/servicios-a-organizaciones/indice-de-transparencia-del-sector-publico-costarricense/datos-abiertos). The dataset contains registered data from 2015 to 2025. The dataset to be downloaded was previously cleaned using OpenRefine.

Demographic and population datasets were sourced from the Costa Rican "Instituto Nacional de Estadística y Censos" (INEC) website [[3]](https://inec.cr/estadisticas-fuentes), while macroeconomic and socioeconomic indicators were retrieved from the World Bank database [[4]](https://data.worldbank.org/country/costa-rica).

### Reading The Crime Dataset

Not all columns will be used for this exploratory data analysis (EDA). Only a few columns will be selected due to inconsistencies in data utility across the years. For example, the exact hour of the registered crime was not recorded in certain years.

```python
import pandas as pd

columns_to_keep = ["Delito", "SubDelito", "Fecha", "Provincia", "Canton"]
crime_data = pd.read_csv("../data/crime_data_2015_2025.csv", usecols=columns_to_keep)
display(crime_data.head(4))
display(crime_data.tail(4))
```

```python
crime_data.info()
```

#### Extracting year and month for each registered crime

```python
crime_data["Fecha"] = pd.to_datetime(crime_data["Fecha"], errors="coerce")

crime_data["Año"] = crime_data["Fecha"].dt.year
crime_data["Mes"] = crime_data["Fecha"].dt.month

crime_data.drop(columns=["Fecha"], inplace=True, errors="ignore")

display(crime_data.head(3))
display(crime_data.tail(3))
```

##### Missing data in the crime dataset
The original data for each year from 2015 to 2025 was retrieved from [[1]](https://sitiooij.poder-judicial.go.cr/index.php/ayuda/servicios-policiales/servicios-a-organizaciones/indice-de-transparencia-del-sector-publico-costarricense/datos-abiertos) and concatenated into a single dataset. However, two years were missing certain months: 2020 was missing November and December, and December was missing from the 2021 data. 

```python
months_sorted = crime_data.groupby("Año")["Mes"].unique().apply(sorted)
print(months_sorted)
```

#### Handling the new datasets for 2020-2022 and 2025

The process will proceed as follows:
1. Read the dataset for each year from the other source [[2]](https://sitiooij.poder-judicial.go.cr/index.php/ayuda/servicios-policiales/servicios-a-organizaciones/indice-de-transparencia-del-sector-publico-costarricense/datos-abiertos).
2. Check consistency across these years to merge the data into a single Dataframe.
3. Perform data wrangling to the new data.
4. Merge the fixed data into the original dataset.


```python
columns_to_keep = ["Delito", "SubDelito", "Fecha", "Provincia", "Canton"]

crimes_2020 = pd.read_csv("../data/crime_data_2020.csv", usecols=columns_to_keep)
crimes_2021 = pd.read_csv("../data/crime_data_2021.csv", usecols=columns_to_keep)
crimes_2022 = pd.read_csv("../data/crime_data_2022.csv", usecols=columns_to_keep)
crimes_2025 = pd.read_csv("../data/crime_data_2025.csv", usecols=columns_to_keep)

display(crimes_2020.head(2))
display(crimes_2020.tail(2))
display(crimes_2021.head(2))
display(crimes_2021.tail(2))
display(crimes_2022.head(2))
display(crimes_2022.tail(2))
display(crimes_2025.head(2))
display(crimes_2025.tail(2))
```

All of the new datasets follow the same structure, so they can be merged into a single DataFrame to simplify the workflow.

```python
fixed_data = pd.concat([crimes_2020, crimes_2021, crimes_2022, crimes_2025], ignore_index=True)
```

```python
fixed_data
```

#### Extracting year and month for each registered crime

```python
fixed_data["Fecha"] = pd.to_datetime(fixed_data["Fecha"], errors="coerce")
fixed_data["Año"] = fixed_data["Fecha"].dt.year
fixed_data["Mes"] = fixed_data["Fecha"].dt.month
fixed_data.drop(columns="Fecha", errors="ignore", inplace=True)
fixed_data
```

#### Verifying complete months for 2020 and 2021

```python
months_sorted = fixed_data.groupby("Año")["Mes"].unique().apply(sorted)
display(months_sorted)
```

All of the new datasets contain data for all months of each year.

#### Verifying the standard crime categorization structure

```python
fixed_data["Delito"].unique()
```

The new dataset share the same crime categorization structure as the other years.

#### Converting text columns to Title Case

```python
columns_to_fix = ["Delito", "SubDelito", "Provincia", "Canton"]
fixed_data[columns_to_fix] = fixed_data[columns_to_fix].map(str.title)
fixed_data
```

### Replacing the original data with the new data

Since the newly corrected dataset and the original dataset share the exact same structure, they can be merged into a single, consistent dataset.

```python
display(crime_data.head(3))
display(fixed_data.head(3))
```

##### Deleting inconsistent crime data for 2020–2022 and 2025

The inconsistent crime data will be deleted from the original dataset to prepare for the merge.

```python
crime_data = crime_data[(crime_data["Año"] != 2020) & (crime_data["Año"] != 2021)
                           & (crime_data["Año"] != 2022)
                           & (crime_data["Año"] != 2025)]
```

```python
crime_data = pd.concat([crime_data, fixed_data], ignore_index=True)
```

```python
display(crime_data.head(3))
display(crime_data.tail(3))
```

#### Validating the final fixes

##### Verifying completed months for each year

```python
months_sorted = crime_data.groupby("Año")["Mes"].unique().apply(sorted)
display(months_sorted)
```

All years now contain the correct number of months.

##### Verifying the crime categorization structure

```python
crime_data["Delito"].unique()
```

The new dataset share the same crime categorization structure as the other years.

#### Translating the dataset into English

To make this project accessible to a global audience, the crime categories and column names will be translated from Spanish to English.

```python
display(crime_data["Delito"].unique())
display(crime_data["SubDelito"].unique())
display(crime_data.columns)
```

All Spanish terms will be translated into English. This process also highlights other errors in the data, such as:
- Incorrect encodings: "Discusion/RiÑ"
- Duplicate subcategories: "Ajuste De Cuentas/venganza" vs. "Ajuste De Cuentas/Venganza"

```python
# Clean encoding issues and trailing spaces first to avoid mapping failures
crime_data['SubDelito'] = crime_data['SubDelito'].str.replace('&#209;', 'Ñ', regex=False)
crime_data['SubDelito'] = crime_data['SubDelito'].str.replace('&#241;', 'ñ', regex=False)
crime_data['SubDelito'] = crime_data['SubDelito'].str.strip()

# Complete dictionary for subcategories (SubDelito)
subcategory_translation = {
    'Descuido': 'Negligence / Left Unattended',
    'Otro O Indeterminado': 'Other or Undetermined',
    'Por Confianza': 'Breach of Trust',
    'Tacha De Vehiculo': 'Vehicle Breaking and Entering',
    'Arma Blanca': 'Edged Weapon / Knife',
    'Arma Contundente': 'Blunt Object',
    'Arma De Fuego': 'Firearm',
    'Arrebato': 'Snatching',
    'Candado Chino': 'Chokehold / Rear-Naked Choke',
    'Desconocido': 'Unknown',
    'Golpes': 'Physical Assault / Blows',
    'Inmovilizacion': 'Immobilization',
    'Intimidacion Verbal': 'Verbal Intimidation',
    'Uso De Gas': 'Use of Pepper Spray / Gas',
    'Ajuste De Cuentas/venganza': 'Score Settling / Revenge',
    'Ajuste De Cuentas/Venganza': 'Score Settling / Revenge',
    'Discusion/riña': 'Altercation / Fight',
    'Discusion/RiÑA': 'Altercation / Fight',
    'Femicidio': 'Femicide',
    'No Determinado': 'Undetermined',
    'Por La Comision De Otro Delito': 'During the Commission of Another Crime',
    'Profesional': 'Professional / Contract Killing',
    'Repeliendo Actividad Criminal': 'Repelling Criminal Activity',
    'Violencia Domestica': 'Domestic Violence',
    'Ardid Previo/distraccion': 'Distraction Scam',
    'Ardid Previo/Distraccion': 'Distraction Scam',
    'Carterista': 'Pickpocketing',
    'Con Llave': 'Using a Key / Lock Picking',
    'Ganzua/varilla': 'Using a Lockpick / Slim Jim',
    'Ganzua/Varilla': 'Using a Lockpick / Slim Jim',
    'Por Descuido': 'Left Unattended',
    'Progresivos': 'Progressive Theft',
    'Retiro De Cajero Automatico': 'ATM Withdrawal Theft',
    'Uso De Somnifero': 'Use of Soporifics / Drugging',
    'Boquete': 'Wall Tunneling / Wall Breach',
    'Corta Candados': 'Bolt Cutter Usage',
    'Escalamiento': 'Climbing / Scaling Walls',
    'Forzadura': 'Forced Entry',
    'Quitan Celosias': 'Louver Window Removal',
    'Ruptura Ventana': 'Window Shattering',
    'Ardid Previo': 'Deceptive Trick / Scam',
    'Asalto': 'Assault',
    'Cocherazo': 'Garage Break-in / Driveway Theft',
    'Ideologico': 'Ideological Motive'
}

# Apply subcategory translation
crime_data['SubDelito'] = crime_data['SubDelito'].map(subcategory_translation).fillna(crime_data['SubDelito'])

# Complete dictionary for main crime categories (Delito)
category_translation = {
    'Robo De Vehiculo': 'Vehicle Theft',
    'Tacha De Vehiculo': 'Vehicle Breaking and Entering',
    'Asalto': 'Assault',
    'Homicidio': 'Homicide',
    'Hurto': 'Theft',
    'Robo': 'Robbery'
}

# Apply category translation
crime_data['Delito'] = crime_data['Delito'].map(category_translation)

# Dictionary for column names
column_translation = {
    'Delito': 'Crime_category',
    'SubDelito': 'Crime_subcategory',
    'Provincia': 'Province',
    'Canton': 'Canton',
    'Año': 'Year',
    'Mes': 'Month'
}

# Rename the columns
crime_data = crime_data.rename(columns=column_translation)
```

```python
crime_data
```

#### Adding missing accents to the provinces of San José and Limón

```python
crime_data["Province"] = crime_data["Province"].replace(
    "San Jose",
    "San José"
)

crime_data["Province"] = crime_data["Province"].replace(
    "Limon",
    "Limón"
)

crime_data
```

The next dataset provides the total population size for each province from 2015 to 2025.

##### Changing dataset indexing
To improve dataset manipulation during the EDA, using an index based on the year is much more effective.

```python
population_by_province = population_by_province.pivot(index="Provincia",columns="Año", values="Población")
population_by_province
```

Only the data from 2015 to 2025 are relevant for this analysis. Therefore, the remaining columns can be dropped.

```python
population_by_province = population_by_province.loc[:,2015:2025]
population_by_province
```
